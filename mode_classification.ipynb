```python
# Load the MIDI file
scores =[]
for i, file in enumerate(files[150:200]):
    print(i, file)
    try:
        scores.append(music21.converter.parse(file))
        print(f'{i} appended')
    except:
        print(f'{i} error')
        continue
len(scores)
```

```python
def extract_pitch_sequence(score):

    # Extract the pitch class sequence
    # Initialize an empty list to store the pitch class sequence
    pitch_class_sequence = []


    # Iterate through all elements in the score
    for element in score.flat:
        # Check if the element is a note
        if 'Note' in element.classes:
            pitch_class_sequence.append(element.pitch.pitchClass)
        # Check if the element is a chord
        elif 'Chord' in element.classes:
            for pitch in element.pitches:
                pitch_class_sequence.append(pitch.pitchClass)
    return pitch_class_sequence
```

```python
pitch_sequences_songs = []
for score in scores:
    pitch_sequences_songs.append(extract_pitch_sequence(score))
```

```python
pitch_class_histogram = []
for indv_pitch_sequence in pitch_sequences_songs:

    # Compute the pitch class histogram
    pitch_class_histogram.append(Counter(indv_pitch_sequence))
pitch_class_histogram
```

```python
def padding_counter(values):
    # calculate the amount of padding needed
    pad_width = 12 - len(values)
    if pad_width >0:
        # pad array a with zeros on the left
        padded_pitch_class = np.pad(values, (pad_width, 0), 'constant', constant_values = (0, 0))
    else:
        padded_pitch_class = np.array(values[:12])
    # check the size of the arrays
    #print(len(padded_pitch_class), padded_pitch_class) # prints 8
    return padded_pitch_class
```

Getting which notes are played the most during the duration of the midi song

```python
padded_class = []
for counter in pitch_class_histogram:
    padded_class.append(padding_counter(list(counter)))
    
len(padded_class), padded_class
```

reference distributions for all 7 modes obtained by Carol L. Krumhansl and George E. Kessler in their study 

"Tracing the dynamic changes in perceived tonal organization in a spatial representation of musical keys"

```python
# Compute the reference distributions for each mode
ionian_reference = [6.35, 2.23, 3.48, 2.33, 4.38, 4.09, 2.52, 5.19, 2.39, 3.66, 2.29, 2.88]
dorian_reference = [6.33, 2.68, 3.52, 5.38, 2.60, 3.53, 2.54, 4.75, 3.98, 2.69, 3.34, 3.17]
phrygian_reference = [6.31, 2.58, 3.82, 2.92, 3.10, 2.88, 4.89, 2.11, 4.54, 3.93, 2.90, 3.07]
lydian_reference = [6.33, 2.68, 3.52, 5.38, 2.60, 3.53, 2.54, 4.75, 3.98, 2.69, 3.34, 3.17]
mixolydian_reference = [6.33, 2.68, 3.52, 5.38, 2.60, 3.53, 2.54, 4.75, 3.98, 2.69, 3.34, 3.17]
aeolian_reference = [6.33, 2.68, 3.52, 5.38, 2.60, 3.53, 2.54, 4.75, 3.98, 2.69, 3.34, 3.17]
locrian_reference = [6.31, 2.58, 3.82, 2.92, 3.10, 2.88, 4.89, 2.11, 4.54, 3.93, 2.90, 3.07]
```

```python
def getting_mode(a):

    # Compute the correlation coefficient for each mode
    ionian_similarity = np.corrcoef(a, ionian_reference)[0, 1]
    dorian_similarity = np.corrcoef(a, dorian_reference)[0, 1]
    phrygian_similarity = np.corrcoef(a, phrygian_reference)[0, 1]
    lydian_similarity = np.corrcoef(a, lydian_reference)[0, 1]
    mixolydian_similarity = np.corrcoef(a, mixolydian_reference)[0, 1]
    aeolian_similarity = np.corrcoef(a, aeolian_reference)[0, 1]
    locrian_similarity = np.corrcoef(a, locrian_reference)[0, 1]


    # Find the mode with the highest similarity score
    mode_similarities = [ionian_similarity, dorian_similarity, phrygian_similarity, lydian_similarity, mixolydian_similarity, aeolian_similarity, locrian_similarity]
    mode_index = mode_similarities.index(max(mode_similarities))
    modes = ['Ionian', 'Dorian', 'Phrygian', 'Lydian', 'Mixolydian', 'Aeolian', 'Locrian']
    mode = modes[mode_index]
    return mode
```

```python
modes = []
for i, arr in enumerate(padded_class):
    modes.append((i, getting_mode(arr)))
    print(modes[-1])
```
